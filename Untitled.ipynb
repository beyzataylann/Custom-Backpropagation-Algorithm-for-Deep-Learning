```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


X = []
Y = []

for i in range(1, 10, 2):
    for m in range(1, 10, 2):
        X.append([i, m])
        Y.append([i * m])

X = np.array(X)

Y = np.array(Y)


X_min = np.min(X, axis=0)
X_max = np.max(X, axis=0)
X_normalized = (X - X_min) / (X_max - X_min)

Y_min = np.min(Y)
Y_max = np.max(Y)
Y_normalized = (Y - Y_min) / (Y_max - Y_min)

class NeuralNetwork(object):
    def __init__(self):
        self.inputSize = 2
        self.outputSize = 1
        self.hiddenSize = 3

        self.W1 = np.random.randn(self.inputSize, self.hiddenSize) * np.sqrt(2. / self.inputSize)
        self.W2 = np.random.randn(self.hiddenSize, self.outputSize) * np.sqrt(2. / self.hiddenSize)

        print(self.W1)
        print(self.W2)
        self.learning_rate = 0.01
        self.bp_history = []

    def feedForward(self, X):

        self.z = np.dot(X, self.W1)
        self.z2 = self.sigmoid(self.z)
        self.z3 = np.dot(self.z2, self.W2)
        output = self.sigmoid(self.z3)
        return output

    def sigmoid(self, s, deriv=False):

        if deriv:
            return s * (1 - s)
        return 1 / (1 + np.exp(-s))

    def backward(self, X, Y, output):

        self.output_error = Y - output
        self.output_delta = self.output_error * self.sigmoid(output, deriv=True)

        self.z2_error = self.output_delta.dot(self.W2.T)
        self.z2_delta = self.z2_error * self.sigmoid(self.z2, deriv=True)


        self.W1 += self.learning_rate * X.T.dot(self.z2_delta)
        self.W2 += self.learning_rate * self.z2.T.dot(self.output_delta)

    def train(self, X, Y, epochs=10000):
        losses = []
        for i in range(epochs):
            output = self.feedForward(X)
            self.backward(X, Y, output)

            if i % 100 == 0:
                loss = np.mean(np.square(Y - output))
                losses.append(loss)
                print(f"Iteration {i}, Loss: {loss}")
        return losses



```

```python


import matplotlib.pyplot as plt
NN = NeuralNetwork()

losses = NN.train(X_normalized, Y_normalized)


predictions_normalized = NN.feedForward(X_normalized)


predictions_denormalized = predictions_normalized * 100
Y_denormalized = Y_normalized * (Y_max - Y_min) + Y_min

plt.figure(figsize=(10, 5))
plt.scatter(Y_denormalized, predictions_denormalized, label='Tahminler', color='blue')
plt.plot([min(Y_denormalized), max(Y_denormalized)], [min(Y_denormalized), max(Y_denormalized)], color='red', label='Regresyon Çizgisi')

plt.title('Gerçek Değerler ve Tahminler')
plt.xlabel('Gerçek Değerler')
plt.ylabel('Tahmin Edilen Değerler')
plt.legend()
plt.show()

```
